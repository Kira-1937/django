```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('census_village_data_at_district_level_Explorin.csv')
df.head(5)
print(df.shape)
df.info()
```

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
df['sex_ratio']=df['total_population_female']/df['total_population_male']
df['primary_school']=df['govt_primary_school']+df['pvt_primary_school']
df['primary_school_per_capita']=df['total_population']/df['primary_school']
df.head()
df.shape
```

```python
df.describe()
```

```python
sns.heatmap(df.isnull())
```

```python
df.head()
```

```python
df.drop(df.iloc[:, :3], inplace = True, axis = 1)
df.head()
```

```python
df.shape
```

```python
df.drop(df.iloc[:, 2:40], inplace = True, axis = 1)
df.head()
```

```python
df.drop('total_households', inplace = True, axis = 1)
df.head()
```

```python
df[(df['sex_ratio']<0.845)&(df['primary_school_per_capita']<1000)]
#so we got 10 district by analysising the village
```

```python
# importing the module
#graph for the district which is selected by analysing the village
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# creating a DataFrame
data_village_selected = {'district' : ['kargil', 'kinnaur', 'dhaulpur', 
                 'west_kemang', 'morena'],  
        'primary_school' : [300, 213,1070 ,146,1757],  
        'sex_ratio' : [0.84,0.82,0.84,0.79,0.83] 
        } 
df_final1 = pd.DataFrame(data_village_selected)
#area graph
df_final1['primary_school'].plot(kind="area",stacked=False,color=['#1970E7','#E73E19'])
```

```python
df_final1["sex_ratio"].plot(kind="area",stacked=False,color=['#1970E7','#E73E19'])
```

```python
#scatter plot between sex_ratio and primary school
df_final1[['sex_ratio','primary_school']].plot(kind ="scatter",x="primary_school",y="sex_ratio")
```

```python
df_final1['sex_ratio'].plot(kind="bar")
```

```python
df_final1['primary_school'].plot(kind="bar")
```
